# Concatenation - by rows

```python
import pandas as pd
```

```python
df1_dummy = {
    "serial_id":["1","2","3","4","5"],
    "sale_month":["Jan","Feb","Mar","Apr","May"],
    "sales":["12300","25100","17800","20100","21000"]
}
```

```python
df1 = pd.DataFrame(df1_dummy,columns=["serial_id","sale_month","sales"])
df1
```

```python
df2_dummy = {
    "serial_id":["6","7","8","9","10"],
    "sale_month":["Jun","Jul","Aug","Sept","Oct"],
    "sales":["25000","23700","24600","24000","23950"]
}
```

```python
df2 = pd.DataFrame(df2_dummy,columns=["serial_id","sale_month","sales"])
df2
```

```python
df3_dummy = {
    "sales_threshold":["No","Yes","No","Yes","Yes","Yes","Yes","Yes","Yes","Yes"],
    "bonus_threshold":["No","Yes","No","No","No","Yes","No","Yes","Yes","No"]
}
```

```python
df3 = pd.DataFrame(df3_dummy,columns=["sales_threshold","bonus_threshold"])
df3
```

```python
df_row = pd.concat([df1,df2],ignore_index=True)
df_row
```

```python
df_full = pd.concat([df_row,df3],axis=1)
df_full
```

# Append

```python
df_append = df1.append(df2)
df_append
```

```python
df_append_full = df_append.append(df3)
df_append_full
```

# Merging - by columns 

one-to-one joins -> same column with same data.,
many-to-one joins -> one column contains duplicate entries.,
many-to-many joins -> both columns contains duplicates

```python
import pandas as pd
```

One-to-One Join

```python
df_1 = pd.DataFrame({"product":["Prod_1","Prod_2","Prod_3","Prod_4"],
                   "division":["Div_A","Div_B","Div_C","Div_B"]
                   })
df_2 = pd.DataFrame({"sales":["12500","10800","5600","7900"],
                   "product":["Prod_3","Prod_2","Prod_4","Prod_1"]
                   })
display(df_1,df_2)
```

```python
df_3 = pd.merge(df_1,df_2)
df_3
```

Many-to-one joins 

```python
df_4 = pd.DataFrame({"division":["Div_A","Div_B","Div_C"],
                     "manager":["Roger","Rafael","Novak"]})
df_4
```

```python
df_5 = pd.merge(df_3,df_4)
df_5
```

```python
df_6 = pd.DataFrame({"division":["Div_A","Div_A","Div_B","Div_C","Div_C","Div_C"],
                     "emp_grade":["13","14+","12","11","10","9-"]})
df_6
```

```python
df_7 = pd.merge(df_1,df_6)
df_7
```

# Merge Keys

On parameter -> explicitly specify the name of the key column.,
Left_on,right_on -> with different column names, but similar data.,
left_index,right_index -> for merging on an index but not a column

```python
 display(pd.merge(df_1,df_2,on="product"))
```

```python
df_1new = pd.DataFrame({"project":["Prod_1","Prod_2","Prod_3","Prod_4"],
                        "division":["Div_A","Div_B","Div_C","Div_B"]})
df_1new
```

```python
df_2
```

```python
display(pd.merge(df_1new,df_2,left_on="project",right_on="product"))
```

```python
display(pd.merge(df_1new,df_2,left_on="project",right_on="product").drop("product",axis=1))
```

```python
df_1index = df_1.set_index("product")
df_2index = df_2.set_index("product")
display(df_1index,df_2index)
```

# How Parameter

```python
df_a = pd.DataFrame({"id":["1","2","3","4"],
                     "actor_first_name":["Robert","Gwyneth","John","Paul"],
                     "actor_last_name":["Downey Jr.","Paltrow","Favreau","Bettany"],
                     "value":["10","6","7","7"]})
df_b = pd.DataFrame({"id":["1","2","3","4","5","6"],
                     "actor_first_name":["Robert","Chris","Chris","Mark","Scarlett","Jeremy"],
                     "actor_last_name":["Downey Jr.","Evans","Hemsworth","Ruffalo","Johansson","Renner"]})
display(df_a,df_b)
```

```python
display(pd.merge(df_a,df_b,how="inner"))    #ONLY COMMON VALUE BETWEEN TWO DATAFRAMES WILL OCCUR
```

```python
display(pd.merge(df_a,df_b,how="left")) #ONLY COMMON VALUE BETWEEN TWO DATAFRAMES AND LEFT DATA WILL OCCUR
```

```python
display(pd.merge(df_a,df_b,how="right")) #ONLY COMMON VALUE BETWEEN TWO DATAFRAMES AND RIGHT DATA WILL OCCUR
```

```python
display(pd.merge(df_a,df_b,how="outer",indicator=True)) #ALL VALUES IN TWO DATAFRAMES DATA WILL OCCUR
```

# Suffixes

To specify customised suffix to avoid confusion (here LD and RD for left and right dataframes) -> To remove ambiguity

```python
df_s1 = pd.DataFrame({"product":["Prod_1","Prod_2","Prod_3","Prod_4"],
                        "division":["Div_A","Div_B","Div_C","Div_D"]})
df_s2 = pd.DataFrame({"product":["Prod_1","Prod_2","Prod_3","Prod_4"],
                        "division":["Div_C","Div_A","Div_B","Div_D"]})
display(pd.merge(df_s1,df_s2,on="product"))
```

```python
display(pd.merge(df_s1,df_s2,on="product",suffixes=["_LD","_RD"]))
```

# Updating Dataframe

To replace or update values of columns in a dataframe from an another dataframe

```python
df_u1 = pd.DataFrame({'c1':['a','a','b','b'],
                      'c2':['x','y','x','y'],
                      'val':0})
df_u2 = pd.DataFrame({'c1':['a','a','b','b'],
                      'c2':['x','y','x','y'],
                      'val':[8,12,15,7]})
display(df_u1,df_u2)
```

```python
df_u1.update(df_u2)
df_u1
```

```python
df_u2.combine_first(df_u1)
df_u2
```
