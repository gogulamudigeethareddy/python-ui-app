```python
import pandas as pd
```

```python
wine_data = pd.read_csv("winequality_red.csv")
wine_data.head()
```

# Simple Feature Scaling

Diving the value with maximum value , Xnew = X/Xmax

```python
wine_data["fixed acidity"] = wine_data["fixed acidity"]/wine_data["fixed acidity"].max()
wine_data.head()    #feature by feature
```

```python
wine_data = wine_data/wine_data.max()      #entire dataset
wine_data.head()
```

# Min-Max Scaling 

Xnew = (X - Xmin)/(Xmax - Xmin)

```python
wine_data["fixed acidity"] = (wine_data["fixed acidity"]-wine_data["fixed acidity"].min())/\
                             (wine_data["fixed acidity"].max()-wine_data["fixed acidity"].min())
wine_data.head()     #feature by feature
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
wine_data[wine_data.columns] = scaler.fit_transform(wine_data[wine_data.columns])
wine_data.head()     #entire dataset
```

# Z-Score Normalization

Ranges from -3 to +3. Also known as standard score. Xnew = (X-mean(μ))/standard deviation(σ)

```python
wine_data["fixed acidity"] = (wine_data["fixed acidity"]-wine_data["fixed acidity"].mean())/\
                             (wine_data["fixed acidity"].std())
wine_data.head()      #feature by feature
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
wine_data[wine_data.columns] = scaler.fit_transform(wine_data[wine_data.columns])
wine_data.head()          #entire dataset
```
