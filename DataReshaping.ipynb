# Pivot Function

Long to Wide...
Index - labels for new frame's index..,
columns - labels for new frame's indes..,
values - for populating new frame's values

```python
import pandas as pd
import numpy as np
```

```python
stocks_df = pd.read_csv("stocks.csv")
stocks_df
```

```python
stocks_pivot = stocks_df.pivot(index="Name",columns="date",values="close")
stocks_pivot
```

```python
stocks_df.pivot(index="Name",columns="date")  #all columns
```

# Melt Function

Wide to long.,
frame->the dataframe to be reshaped.,
Id_vars->columns to use as identifier variables.,
value_vars->columns to unpivot

```python
addresses_df = pd.read_csv('us-500.csv')
addresses_df
```

```python
addresses_melt = pd.melt(addresses_df,id_vars=["first_name"])
addresses_melt            #wide to long
```

```python
pd.melt(addresses_df,id_vars=["first_name"],value_vars=['city'])  #few columns
```

```python
#customize variable and value names
pd.melt(addresses_df,id_vars=['first_name'],value_vars=['city'],var_name='What Variable',value_name='Which City?')
```

# Stack and Unstack Function

stack->wide to long,unstack->long to wide
used in multi-indexed dataframe
level parameter indicates the level at which stacking/unstacking occurs 

```python
#creating a multi index dataframe
header = pd.MultiIndex.from_product([['Before Upskilling','After Upskilling'],['Age','Salary']])
data = ([[25,45000,27,56000],[38,89000,41,102000],[30,67000,31,70000],[45,130000,48,160000]])

upskilling_df = pd.DataFrame(data,index=['Mark','Anthony','John','Daniel'],columns=header)
upskilling_df
```

```python
upskill_stacked = upskilling_df.stack()
upskill_stacked             #stacking
```

```python
upskilling_df.stack(level=0)      #level 0 stacking
```

```python
upskill_stacked.unstack()    #unstack
```

# Pivot Table

only works with numeric values,non-numeric values will be dropped
mean is default aggregation function,can be changed to others

```python
stocks_pivottable = pd.pivot_table(stocks_df,index="Name")
stocks_pivottable    #all pivot tables require atleast one index
```

```python
pd.pivot_table(stocks_df,index=["Name","date"])
```

```python
pd.pivot_table(stocks_df,index="Name",aggfunc=np.sum,values="volume")
```

```python
pd.pivot_table(stocks_df,index="Name",aggfunc=[np.amin,np.amax])
```
