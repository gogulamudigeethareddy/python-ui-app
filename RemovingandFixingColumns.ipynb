```python
import pandas as pd
```

# Dropping Columns

```python
data = pd.read_csv("tate.csv")
data.head()
```

```python
data.drop(0)     #dropped 1st row
```

```python
data.drop("artist.gender",axis=1)    # to drop columns,axis should be mentioned
```

```python
data.drop(columns=["artist.gender"])     #another method to drop column
```

```python
data.drop(columns=['artist.gender','artist.role']) #to remove more no.of columns
```

```python
data.drop(labels=[0,1,2])  #to drop rows
```

```python
data         #no change in the data
```

```python
data.drop(columns=['artist.role'],inplace=True)   #will remove in original data
```

```python
data       
```

```python
data = pd.read_csv('tate.csv',usecols=['artist.name','data.title'])    #without dropping multiple columns
data
```

# Change Column Casing

```python
data = pd.read_csv('tate.csv')
```

```python
data.columns
```

```python
data.columns.str.upper()
```

```python
[x.upper() for x in data.columns]           
```

```python
data.columns = [x.upper() for x in data.columns]      #original data will be changed
```

```python
data.columns
```

```python
data = pd.read_csv('tate.csv')
```

```python
data.columns = map(lambda x:x.upper(),data.columns)    #'map' can also be used
```

```python
data.columns
```

```python
data = pd.read_csv('tate.csv')
```

```python
import re
```

```python
data.columns = [re.sub(r'([A-Z])',r'_\l', x).lower() for x in data.columns]   #CHECK AGAIN  #underscore will be added when there is a capital letter in the middle of the word.
```

# Rename Columns

```python
data = pd.read_csv('tate.csv')
```

```python
data.columns
```

```python
data.rename(columns={"metadata.acquisition date":"metadata.acquisition year"})
```

```python
data.rename({"artist.gender":"gender"},axis=1)    #another way to rename a column
```

```python
data
```

```python
data.rename(columns={"metadata.acquisition date":"metadata.acquisition year"},inplace=True)
```

```python
data
```

```python
data.rename(columns=lambda x:x.upper(),inplace=True)   #another way to rename a columns
```
