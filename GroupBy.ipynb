# Grouping -> Split-Apply-Combine

Splitting into groups based on a criteria,Apply indicates a functiion or an action like sum, mean etc.,and then combine into single dataset in a sorted way.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic["Pclass"].nunique()
```

```python
titanic["Survived"].sum()
```

```python
titanic_class = titanic.groupby("Pclass")     #Group based on Pclass
titanic_class
```

```python
for i,j in titanic_class:
    print(i)
    print(j)
```

```python
titanic_class.get_group(1)   #Pclass 1 group
```

Aggregate

```python
titanic_class.sum()
```

```python
titanic_class.mean()         #Average age
```

```python
titanic_class.max()  #max fare
```

```python
titanic_class["Fare"].agg({"Fare":['sum','max']})  #check #multiple functions
```

Filtration

```python
#average fare if the average age of group is > or < 38
```

```python
titanic_class.filter(lambda x : x["Age"].mean()>38.0)
```

```python
titanic_class.filter(lambda x : x["Age"].mean()<38.0)
```

```python
titanic_class.filter(lambda x : x["Age"].mean()>38.0)["Fare"].mean()
```

```python
titanic_class.filter(lambda x : x["Age"].mean()<38.0)["Fare"].mean()
```

Transformation

```python
#fare of the passenger higher than average for their ticket class
```

```python
titanic["avg_fare_class"] = titanic.groupby("Pclass")["Fare"].transform(lambda x : x.mean())
```

```python
titanic.head()
```

```python
titanic["fare_above_avg"] = titanic["avg_fare_class"] < titanic["Fare"]
titanic
```

```python
titanic.head()
```

```python
titanic.loc[:2]
```

```python
titanic_multi = titanic.set_index(["Embarked","Sex"])  #setting name and age as indices
titanic_multi.head()
```

```python
titanic_df = titanic_multi.groupby(["Pclass","Sex"]).mean()  #Group using two columns
titanic_df 
```
