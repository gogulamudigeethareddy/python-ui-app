# Replacing BadData

```python
data = pd.read_csv("tate.csv",low_memory=False)
data
```

```python
pd.isna(data.loc[:,'artist.death.location'])   #checking for na values
```

```python
from numpy import nan
```

```python
data.replace({'artist.death.location':{'Not known':nan}},inplace=True)     #replacing with na inplace of any value 
```

```python
data.head()
```

```python
data = pd.read_csv("tate.csv",low_memory=False)
data
```

```python
data.loc[data['artist.death.location'] == 'not known',['artist.death.location']] = nan   #another way of replacing
```

```python
data.head()
```

```python
data.loc[data['metadata.creation year'].notnull() & data['metadata.creation year'].astype(str).str.contains('[^0-9]')]    
# If found any bad years(like string values), they can be found here
```

```python
data.loc[data['metadata.creation year'].notnull() & data['metadata.creation year'].astype(str).str.contains('[^0-9]'),['metadata.creation year']] = nan
#converted bad values to nan 
```

# Identify and Drop Duplicates

```python
data = pd.read_csv("tate.csv")
```

```python
data.head()
```

```python
data.drop_duplicates()    #Drops the rows if two/more rows are same with all values
```

```python
data.drop_duplicates(subset = ['data.title'])  #removed the rows with same name in the title column
```

```python
data.drop_duplicates(subset = ['data.title'],keep='first')   #keeps 1st row and drops the last row with same data 
```

```python
data.drop_duplicates(subset = ['data.title'],keep='last',inplace=True)   #keeps 1st row and drops the last row with same data 
```

```python
data.drop_duplicates(subset = ['data.title'],keep=False)   #Does not keep the none of the duplicated data
```

```python
data = pd.read_csv('tate.csv',low_memory=False)
```

```python
data.shape
```

```python
data.duplicated()
```

```python
data.loc[data.duplicated()]   #not fully duplicated
```

```python
data.duplicated(subset = ['data.title'],keep=False)    #to find duplicates
```

```python
data.loc[data.duplicated(subset = ['data.title'],keep=False)]     #If'Not known'in the column->not fully duplicates
```
