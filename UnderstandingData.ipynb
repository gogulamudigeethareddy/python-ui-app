# Normalize Data

```python
data['dimensions.height'].dtype  
```

```python
data['dimensions.height'].mean()
```

```python
data['dimensions.height'].min()
```

```python
data['dimensions.height'].max()
```

```python
data['dimensions.height'].std()
```

```python
height = data['dimensions.height']
```

```python
norm_1 = (height-height.mean())/height.std()
norm_1
```

```python
minmax = (height-height.min())/(height.max()-height.min())
minmax        #normalizing between 0 and 1
```

```python
minmax.min()
```

```python
minmax.max()
```

```python

data #no change in dataframe
```

```python
data['dimensions.height'] = minmax
```

```python
data['standardized_height'] = norm_1
```

```python
data_new_column = norm_1
```
