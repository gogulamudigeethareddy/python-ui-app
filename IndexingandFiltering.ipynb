# str.contains

```python
data = pd.read_csv("tate.csv")
```

```python
data.head()
```

```python
data['data.medium']
```

```python
type(data['data.medium'])
```

```python
data['data.medium'].str
```

```python
data['data.medium'].str.contains('Graphite')
```

```python
data.loc[data['data.medium'].str.contains('Graphite'),['artist.name','data.medium']]
```

```python
data.head()     #some of them can be missed.,because of case sensitive
```

```python
data.loc[data['data.medium'].str.contains('Graphite',case=False),['artist.name','data.medium']]
```

```python
data.loc[data['data.medium'].str.contains('(?i)Graphite',regex=True),['artist.name','data.medium']]
```

```python
data.loc[data['data.medium'].str.contains('(?i)Graphite',regex=True) | data['data.medium'].str.contains('(?i)Line',regex=True),['artist.name','data.medium']]
```

```python
data.loc[data['data.medium'].str.contains('(?i)Graphite',case=False,regex=True),['artist.name','data.medium']]   #case insensitive
```

```python
data.dtypes
```

```python
data['metadata.creation year'].str.contains('1826')   #As it is a float data type
```

```python
data["metadata.creation year"].astype(str).str.contains('1926')
```

```python
data.loc[data['metadata.creation year'].astype(str).str.contains('1926')]
```

```python
data.loc[data['artist.death.location'].str.contains('United Kingdom')]   #Contains nan values
```

```python
data.loc[data['artist.death.location'].str.contains('United Kingdom',na=False)] #'na' is assigned to 'False'
```
