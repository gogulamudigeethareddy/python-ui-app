```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('forestfires.csv')
data.head()
```

```python
#Mean

from statistics import mean

a1 = data['temp'].mean()   #using pandas function
a2 = mean(data['temp'])      #statistics function
a3 = np.mean(data['temp'])    #numpy mean
print(a1,a2,a3)
```

```python
#Mode

from statistics import mean
from scipy import stats

a1 = data['temp'].mode()   #using pandas function
#a2 = mode(data['temp'])      #statistics function
a3 = stats.mode(data['temp'])    #stats function.,numpy doesnot have mode
print(a1,a3)
```

```python
#Median

from statistics import median

a1 = data['temp'].median()   #using pandas function
a2 = median(data['temp'])      #statistics function
a3 = np.median(data['temp'])    #numpy function
print(a1,a2,a3)
```

```python
#Standard Deviation

from statistics import stdev  
a1 = data['temp'].std()   #using pandas function
a2 = stdev(data['temp'])      #statistics function
a3 = np.std(data['temp'])    #numpy function
print(a1,a2,a3)
```

```python
#Variance

from statistics import variance  
a1 = data['temp'].var()   #using pandas function
a2 = variance(data['temp'])      #statistics function
a3 = np.var(data['temp'])    #numpy function
print(a1,a2,a3)
```

```python
#Skewness

from scipy.stats import skew 
a1 = data['temp'].skew()   #using pandas function
a2 = skew(data['temp'])      #statistics function
print(a1,a2)
```

```python
#Kurtosis

from scipy.stats import kurtosis

a1 = data['temp'].kurt()   #using pandas function
a2 = kurtosis(data['temp'])      #statistics function
print(a1,a2)
```

Proportion = Eigen value/Sum of Eigen values.,then cumulative proportion is calculated.,

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
tips.head()
```

```python
fig = plt.figure(figsize=(10.5,7.2))
ax = sns.scatterplot(x='total_bill',y='tip',data=tips)

plt.axis('equal')         #Scatterplot with Equalized axis
```

```python
tips_array = tips[['total_bill','tip']].values
tips_array = StandardScaler().fit_transform(tips_array)
```

```python
pca = PCA(n_components=2)
pca.fit(tips_array)
```

```python
print(f'PCA Components: \n {pca.components_} \n')        #to represent directions,variance
print(f'PCA Explained Variance: \n {pca.explained_variance_} \n')
print(f'PCA Explained Variance Ratio: \n{pca.explained_variance_ratio_} \n')
```

```python
fig =plt.figure(figsize=(10,7.5))
sns.scatterplot(x=tips_array[:,0],y=tips_array[:,1])

for length,vector in zip(pca.explained_variance_,pca.components_):
    v = vector*3*length
    ax = plt.gca()
    ax.annotate('',pca.mean_+v,pca.mean_,arrowprops=dict(arrowstyle='-|>',
                                                         linewidth=2,
                                                         shrinkA=0,
                                                         shrinkB=0))
    
plt.axis('equal')
```

between variables..,
Types-> Exploratory(assumes any observed variable is associated with any factor) and Confirmatory(assumes each factor 
is associated with certain subset of observed variables.)..,

# clustering

Divide into groups..,Types->Hierarchical,Non-hierarchical and Model based

```python
from sklearn import datasets
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
```

```python
iris_df = datasets.load_iris()

x = iris_df.data
y = iris_df.target
```

```python
fig = plt.figure(figsize=(10,7.5))

plt.scatter(x[:,0],x[:,1],c=y,s=100)
plt.xlabel('Sepal Length',fontsize=14)
plt.ylabel('Sepal Width',fontsize=14)
```

```python
#K-Means Clustering

mean_dist = []
max_k = 10

for k in range(1,max_k):
    km = KMeans(n_clusters=k,random_state=42)
    km.fit(x)
    mean_dist.append(sum(np.min(cdist(x,km.cluster_centers_,'euclidean'),axis=1))/x.shape[0])
fig = plt.figure(figsize=(10,7.5))
plt.plot(range(1,max_k),mean_dist)
```

```python
km = KMeans(n_clusters=3,random_state=42)
km.fit(x)
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,7.5))

ax[0].scatter(x[:,0],x[:,1],c=y,s=100)
ax[0].set_xlabel('Sepal Length',fontsize=14)
ax[0].set_ylabel('Sepal Width',fontsize=14)
ax[0].set_title('Real',fontsize=14)

ax[1].scatter(x[:,0],x[:,1],c=km.labels_,s=100)
ax[1].set_xlabel('Sepal Length',fontsize=14)
ax[1].set_xlabel('Sepal Length',fontsize=14)
ax[1].set_xlabel('Predicted',fontsize=14)
```

```python
#Hierarchical Clustering

from scipy.cluster.hierarchy import dendrogram,linkage
L = linkage(x,'ward')
```

```python
plt.figure(figsize=(15,8))

dendrogram(L,truncate_mode='lastp',p=100)
plt.axhline(y=10,c='k')
```

```python
from sklearn.datasets import load_diabetes

db = load_diabetes()   #Each of these 10 feature variables have been mean centered and scaled by standard deviation times

df = pd.DataFrame(db.data)
df.columns = db.feature_names
df['target'] = db.target
df.head()
```

```python
df.isnull().sum()
```

```python
df['age'][3] = np.nan
df.head()
```

```python
df.isnull().mean()
```

```python
#Imputation

df.columns[df.isnull().mean()>0]
```

```python
threshold = 0.001

#dropping columns with missing value rate higher than threshold

data = df[df.columns[df.isnull().mean()<threshold]]
data.head()
```

```python
#dropping rows with missing value rate higher than threshold

data = df.loc[df.isnull().mean(axis=1) < threshold]
df.head()
```

```python
#filling all missing values with 0

data = df.fillna(0)
data.head()
```

```python
#filling missing values with medians of the columns

data = df.fillna(df.median())
data.head()
```

```python
from sklearn.impute import SimpleImputer

i_mean = SimpleImputer(missing_values=np.nan,strategy='mean')
i_mean.fit(df)

print(f'Mean {df["age"].mean()}')

imputed = pd.DataFrame(i_mean.transform(df))
imputed.head()
```

```python
#Outliers

#Standard Deviation policy
upper = df['age'].mean()+3*df['age'].std()
lower = df['age'].mean()-3*df['age'].std()

data = df[(df['age']<upper) & (df['age']>lower)]
data.head()
```

```python
#Percentile Deviation Policy
upper = df['age'].quantile(.75)
lower = df['age'].quantile(.25)

IQR = upper-lower   #Interquartile range

data = df[(df['age']<upper+1.5*IQR) & (df['age']>lower-1.5*IQR)]
data.head()
```

```python
#Binning
#It helps preventing overfitting but sacrifice performance(use with caution!)
df['binned_age'] = pd.cut(df['age'],
                          bins=[0,0.01,0.02,0.05,0.09],
                          labels=['low','mid-low','mid-high','high'])
df.head()
```
